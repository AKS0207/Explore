### Preparing Data

#### Following Columns can be dropped from the dataframe

- 'instant' column can be dropped as it is just a record indexing column
- 'dteday' column can be dropped as we dont require the date of record for our analysis
- 'registered' and 'casual' column can also be removed as our target variable is the 'cnt' column of total count

```python
bike = bike.drop('instant',axis=1)
bike = bike.drop('dteday',axis=1)
bike = bike.drop('registered',axis=1)
bike = bike.drop('casual',axis=1)
bike.head()
```

#### Converting numerical values of column 'season' into categorical string values

```python
season_string = {'season':{1:'spring', 2:'summer', 3:'fall', 4:'winter'}}
bike.replace(season_string,inplace=True)
bike.head()
```

### Dummy Variables

```python
# Getting the dummy variables for the column 'season' and storing it in a new variable - 'status'

status = pd.get_dummies(bike['season'], drop_first=True)
status.head()
```

```python
# Adding the results to the original bike dataframe & dropping 'season' as we have created the dummies for it

bike = pd.concat([bike, status], axis=1)
bike.head()
```

```python
bike = bike.drop('season', axis=1)
bike.head()
```

```python
# Getting the dummy variables for the column 'weathersit' and storing it in a new variable - 'status1'

status1 = pd.get_dummies(bike['weathersit'], drop_first=True)
status1 = status1.rename(columns={ 1: 'Clear,FewClouds, PartlyCloudy', 
                                  2: 'Mist+Cloudy,Mist+BrokenClouds,Mist+FewClouds,Mist',
                                  3: 'LightSnow,LightRain+Thunderstorm+ScatteredClouds,LightRain+ScatteredClouds',
                                  4: 'HeavyRain+IcePallets+Thunderstorm+Mist,Snow+Fog'})
status1.head()
```

```python
# Adding the results to the original bike dataframe & dropping 'weathersit' as we have created the dummies for it

bike = pd.concat([bike, status1], axis=1)
bike = bike.drop('weathersit', axis=1)
bike.head()
```

```python
# Getting the dummy variables for the column 'mnth' and storing it in a new variable - 'status2'

status2 = pd.get_dummies(bike['mnth'], drop_first=True)
status2 = status2.rename(columns={ 1: 'jan', 2: 'feb', 3: 'mar', 4: 'apr', 5: 'may', 6: 'jun', 7: 'jul', 8: 'aug', 9: 'sep', 10: 'oct', 11: 'nov', 12: 'dec'})
status2.head()
```

```python
# Adding the results to the original bike dataframe & dropping 'mnth' as we have created the dummies for it

bike = pd.concat([bike, status2], axis=1)
bike = bike.drop('mnth', axis=1)
bike.head()
```

```python
# Getting the dummy variables for the column 'weekday' and storing it in a new variable - 'status3'

status3 = pd.get_dummies(bike['weekday'], drop_first=True)
status3 = status3.rename(columns={ 1: 'mon', 2: 'tue', 3: 'wed', 4: 'thu', 5: 'fri', 6: 'sat', 7: 'sun'})
status3.head()
```

```python
# Adding the results to the original bike dataframe & dropping ''weekday' as we have created the dummies for it

bike = pd.concat([bike, status3], axis=1)
bike = bike.drop('weekday', axis=1)
bike.head()
```

### Dividing into X and Y sets for the model building

```python
y_train = bike_train.pop('cnt')
X_train = bike_train
```

### Building a linear model

We will be using the LinearRegression function from SciKit Learn for its compatibility with RFE (which is a utility from sklearn)

### RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_train dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor( X_train_rfe.values, i) for i in range( X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Building model using statsmodel, for the detailed statistics

```python
# Adding a constant variable
import statsmodels.api as sm

X_train_lm1 = sm.add_constant(X_train_rfe)

# Running the linear model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
lm.params
```

```python
# Summary of our linear model
print(lm.summary())
```
